```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/covid_data.csv')

print("Dataset Shape:", df.shape)
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
total_cases = df["Confirmed"].sum()
print("Total Confirmed Cases:", total_cases)
```

```python
total_deaths = df["Deaths"].sum()
print("Total Deaths:", total_deaths)
```

```python
country_cases = df.groupby("Country")["Confirmed"].max()

print("Country with highest cases:")
print(country_cases.idxmax())
print("Total cases:", country_cases.max())
```

```python
country_deaths = df.groupby("Country")["Deaths"].max()

print("Country with highest deaths:")
print(country_deaths.idxmax())
print("Total deaths:", country_deaths.max())
```

```python
avg_cases = np.mean(df["Confirmed"])
print("Average Confirmed Cases:", avg_cases)
```

```python
max_cases = np.max(df["Confirmed"])
print("Maximum Confirmed Cases:", max_cases)
```

```python
min_cases = np.min(df["Confirmed"])
print("Minimum Confirmed Cases:", min_cases)
```

```python
top_cases = df.groupby("Country")["Confirmed"].max().sort_values(ascending=False).head(10)

print("Top 10 Countries with Highest Cases:")
print(top_cases)
```

```python
top_deaths = df.groupby("Country")["Deaths"].max().sort_values(ascending=False).head(10)

print("Top 10 Countries with Highest Deaths:")
print(top_deaths)
```

```python
df["New_Cases"] = df.groupby("Country")["Confirmed"].diff()

print(df[["Date", "Country", "Confirmed", "New_Cases"]].head(10))
```
