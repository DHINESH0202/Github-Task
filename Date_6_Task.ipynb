```python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset("tips")

print("Dataset Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Rows before removing duplicates:", df.shape)

df = df.drop_duplicates()

print("Rows after removing duplicates:", df.shape)
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()

print(df.columns)
```

```python
high_tip = df[df['tip'] > 5]
print(high_tip.head())
```

```python
high_bill = df[df['total_bill'] > 30]
print(high_bill.head())
```

```python
female_customers = df[df['sex'] == 'Female']
print(female_customers.head())
```

```python
sorted_bill = df.sort_values(by='total_bill', ascending=False)
print(sorted_bill.head())
```

```python
sorted_tip = df.sort_values(by='tip', ascending=False)
print(sorted_tip.head())
```

```python
avg_bill = np.mean(df['total_bill'])
print("Average Total Bill:", avg_bill)
```

```python
min_bill = np.min(df['total_bill'])
print("Minimum Bill:", min_bill)
```

```python
total_tip = np.sum(df['tip'])
print("Total Tips Collected:", total_tip)
```

```python
avg_tip_gender = df.groupby('sex')['tip'].mean()
print(avg_tip_gender)
```

```python
bill_by_day = df.groupby('day')['total_bill'].sum()
print(bill_by_day)
```

```python
tip_by_day = df.groupby('day')['tip'].mean()
print(tip_by_day)
```

```python
customers_by_time = df.groupby('time')['size'].sum()
print(customers_by_time)
```

```python
df.to_csv("tips_cleaned_dataset.csv", index=False)
```
